```python
# STEP: specify the censsu Search Criteria and will output the plotable corlation plots
# Using censusdata API:  https://pypi.org/project/CensusData/. 
# GOAL - DASHBOARD 3: Adding new census tract data to 2017-2018 season csv and outputing to file. dashboard3.py will import file and run it in dash
import pandas as pd
import censusdata
import numpy as np
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.precision', 2)
import statsmodels.formula.api as sm

from scipy import stats
import plotly.express as px            #plotting
```

```python
# PARAMETER OF SEARCH 
searchCriteria = 'percentage'           # The type of data to get
startEndRow = [0,300]                   # Range of data variables to import
season_to_show = "ALL"            # Season to plot options: Summer, Autumn, Winter, Spring, ALL_SEASONS

sesonalCSVFile = "DataFiles/FDNY/Season2017_18.csv"
```

```python
 # 1) Find a table by looking at the variables of all tables
table = censusdata.search('acs5', 2018, 'concept', searchCriteria)[startEndRow[0]:startEndRow[1]]
tableColDict = dict()
tableColCodeArray = []
for i in range(0,len(table)):  
    tableColCode = str(table[i]).split(", ")[0].strip("(").strip(")").strip("'")
    tableColSection = str(table[i]).split(", ")[1].strip("(").strip(")").strip("'")
    tableColVar = str(table[i]).split(", ")[2].strip("(").strip(")").strip("'")
    
    tableColDict.update( {tableColCode: [tableColSection,tableColVar]} )
    tableColCodeArray.append(tableColCode)
count = 0
for key in tableColDict:
#     print(str(count)+  "..."+ 
    print(key + "\n              " + tableColDict[key][0] +  "\n              " + tableColDict[key][1])
    count = count + 1
    if count == 101:
        break
keys = list(tableColDict.keys())
if (len(table)<=0):
    print("RETRIEVED NOTHING INPUT NAOTHER SEARCH CRITERIA")
    exit()
```

```python
# 3) Using the target col varables, download em
df = censusdata.download(
    'acs5', 2015,
    censusdata.censusgeo([('state', '36'),('tract', '*')]),#('block group', '*')]),
#     ['B23025_003E', 'B23025_005E', 'B15003_001E', 'B15003_002E', 'B15003_003E',
#       'B15003_004E', 'B15003_005E', 'B15003_006E', 'B15003_007E', 'B15003_008E',
#       'B15003_009E', 'B15003_010E', 'B15003_011E', 'B15003_012E', 'B15003_013E',
#       'B15003_014E', 'B15003_015E', 'B15003_016E'
#     ]
    tableColCodeArray
    
)
df = df.reset_index()
df = df.rename(columns={'index':'Geoid'})
df
```

```python
# Merge into GEOID col and make a CountyName Col
df["CountyName"] = np.str 

for row in range(0, len(df)):
    splitTractName = str(df.iloc[row]["Geoid"]).split(',')

    countyName = splitTractName[1]
    
    geoidSplit = splitTractName[3].split("> ")
    geoid = int(geoidSplit[0].split(':')[1] + geoidSplit[1].split(':')[1] + geoidSplit[2].split(':')[1] )
    
    df.at[row, 'Geoid'] = geoid
    df.at[row, 'CountyName'] = countyName
df
```

```python
# 1) Turning 'Nones' to -1. Del Cols with 400+ 'Nones'
cols = list(df.columns)
# for col in cols:
#     if col == 'Geoid':
#         continue
#     count = 0
#     for row in range(0, len(df)):
#         if df.iloc[0][col] == "None":
#             count = count + 1
#             df.at[row, col] = -1
# #         if count >= 400:
# #             del df[col]
# #             continue
# df
```

```python
# Rename Cols:
for col in cols:
    if col in keys:
        newCol = "("+str(tableColDict[col][0])+") "+str(tableColDict[col][1])
        df = df.rename(columns={col:newCol})
df
```

```python
# PART B: FIND CORRELATION

# 1) ADD THE DEMOGRAPHICS COLS TO THE SESONAL 2017 AND 2018 CSV:
seasonalDF  = pd.read_csv(sesonalCSVFile)  
seasons = ["Spring","Summer","Autumn","Winter"]
demoDF  = df.copy()

# Sum up the count of the seasonal report counts and make a new col:
allSum = []
for row in range(0, len(seasonalDF)):
    sum = 0
    for season in seasons:
        sum = sum + seasonalDF.iloc[row][season]
    allSum.append(sum)
seasonalDF["ALL_SEASONS"] = allSum
seasonalDF
```

```python
# 2) Merge (****output csv here)
GEOID_COL = "Geoid"
demoDF = demoDF.rename(columns={"Id2": GEOID_COL})
seasonalDemoDF = pd.merge(seasonalDF, demoDF, on=[GEOID_COL], how='left') 
# 3) Get the DATA:
seasonalDemoDF.set_index(GEOID_COL, inplace=True)     # setting index for correlation
seasonalDemoDF
```

```python
# 4) PEARSON R CORRELATION:
corr = seasonalDemoDF.corr(method='pearson')
corr.index.name = 'DemographicVariable'                          # name the index column
corr = corr.reset_index()                                        # numbering the rows and setting DemographicVariable as a new coumn
corr
```

```python
# 5) Getting rid of the repeated season rows (they are repeated in the columns): (***output csv after corr)
for season in seasons:
    corr = corr[corr.DemographicVariable != season] 
corr = corr[corr.DemographicVariable != "ALL_SEASONS"] 

# Need to delete the repeated demographic variable columns (they are in the rows already)
delHeader = list(corr.columns)
for i in range(0, len(seasons)):
    delHeader.remove(seasons[i])
delHeader.remove("DemographicVariable")
delHeader.remove("ALL_SEASONS")

for i in range(0, len(delHeader)):
    corr = corr.drop(delHeader[i], axis=1)
for i in delHeader:
    print(i)
corr = corr.reset_index()
corr = corr.drop(['index'], axis=1)              # i have two indexes, so droping old one
corr
```

```python
# 6) PLOT
if season_to_show == "ALL":
    season_to_show = "ALL_SEASONS"
    fig = px.scatter(corr, x="DemographicVariable", y=season_to_show, color=season_to_show, 
                width=900, height=700)
    fig.update_layout(
        title = '(Pearson r Correlation) Reports of All Season VS Picked Demographic Data',
        yaxis_title = "Correlation between Reports per Census Tract (All Seasons)"+' vs Demographic Variables',
        xaxis_title = "Demographic Variables",
    )
fig.show()
```
