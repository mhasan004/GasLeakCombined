```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px

crimeCSVFILE = "NYPD_Complaint_Data_Historic_WITH_CENSUS_DATA_Filtered_Merged_FilteredMore.csv"
df = pd.read_csv(crimeCSVFILE)

colorCol = "BoroughColors"
countyColorOrder = ["Richmond County", "Queens County", "Bronx County","Kings County","New York County"]
GEOID_to_plot = "ALL"
filteredCol = [
    'Geoid',
    'TotalCrime',
    'Spring',
    'Summer',
    'Autumn',
    'Winter',
    'ALL_SEASONS',
    'CountyName',
    'CountyName',
    'CensusTract_2010_NAME',
    'BoroughColors'
]

df
```

```python
# a) Possible rnage of tick bar. needs to be size of df to get lal geoids
def pick_geoid(pick = "ALL"):
    if pick == "ALL":
        tick_tickIds = list(range(0,len(df)))        # [0...2024] - tickvals, values. df row ids for all stuff im printign
        tick_geoidTickVals = list(df["Geoid"])       # will be tickvals and values
        tick_geoidTickVals_str = [ str(x) for x in tick_geoidTickVals ] #turnign geoidList to geoidList where the string geoid and ints
        return [tick_tickIds, tick_geoidTickVals_str]
    tick_tickIds = df.index[df['Geoid'] ==int(pick)].tolist() #the df row id for that geoid
    tick_geoidTickVals_str = [str(tick_tickIds)]
    return [tick_tickIds, tick_geoidTickVals_str]
```

```python
df = df[filteredCol]
df
```

```python
GEOID_to_plot = "ALL",#df.iloc[0]["Geoid"]
# 2) Making an array of dictionaries. Each dict is holds the attributes of each bar 
dimList = list()
dimList.append(
    dict(
        range    = [0, len(df)],                                          # 1) tick0, tick1, etc....
        tickvals = list(range(0,len(df), 100)),  # (1) numbering ticks to be named 
        values   = pick_geoid(GEOID_to_plot)[0],                 # (2) which rows to shows? selects the tickids
        label    = 'Geoid',                      # (3) label the bar
        ticktext = pick_geoid(GEOID_to_plot)[1],       # (4) name of each tick?
    ))              
for col in df:
    if col == "Geoid" or col == "CountyName" or col == "CensusTract_2010_NAME" or col == "BoroughColors":
        continue
#     if col == "Spring":
    parallelBarForCol = dict(
        range = [df[col].min(), df[col].max()],
        label = col, 
        values = df[col]
    )
    dimList.append(parallelBarForCol)
```

```python
# 3) Plot:
figParallel = go.Figure(
    data = go.Parcoords(
        # LINE COLOR DEF: 
        line = dict( 
           color = df[colorCol], #each lines' color val
           colorscale = ["purple", "lightcoral", "red", "firebrick","maroon"], #[counties0->4]
           colorbar = dict(
               tickvals = [0,1,2,3,4],                                       #[counties0->4]
               ticktext = countyColorOrder,                                   #[counties0->4]
               title = {'text': "NYC Counties"},
               x = -0.5,      #******delete this to turn the colorscale bar to normal position
               ticks = "outside",
           ),
           showscale = True, #show the scale 
           cmin = df[colorCol].min(),
           cmax = df[colorCol].max(),
        ),
        dimensions = dimList,
        labelangle = -45,
    )
)

figParallel.update_layout(
    title = "Focused look on Season Gas Leak Reports and Total Crime of all GEOID in NYC",
    autosize=False,
    width  = 900,
    height = 600,
    margin=dict(
        l=10,
        r=100,
        b=10,
        t=200,
        pad=4
    ),
    
)

# figParallel.update_yaxes(automargin=True)
figParallel.show()
```
