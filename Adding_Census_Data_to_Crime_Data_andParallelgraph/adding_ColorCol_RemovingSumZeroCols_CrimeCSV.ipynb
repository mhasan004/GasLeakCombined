```python
# Changing the file to add a color column and changed the totalCrime col to the beginning. Also delete cols that had a sum of zer0
import pandas as pd

crimeCSVFILE = "NYPD_Complaint_Data_Historic_WITH_CENSUS_DATA_Filtered_Merged.csv"
crimeCSVFILE2 = "NYPD_Complaint_Data_Historic_WITH_CENSUS_DATA_Filtered_Merged_FilteredMore.csv"

df = pd.read_csv(crimeCSVFILE)
df
```

```python
# 2) making a dictionary for the different countys
bSet = set(df["CountyName"])
bKey = dict()
count = 0
for borough in bSet: 
    bKey[borough] = count 
    count = count+1
bKey
```

```python
# 3) Adding county number to the rows
for row in range(0,len(df)):
    df.at[row, "BoroughColors"] = bKey[df.loc[row]["CountyName"]]
df
```

```python
# df.to_csv(crimeCSVFILE, index=False)
```

```python
# 4) Deleting cols where sum = 0
deleteCols = []
for col in df:
    if df[col].sum() == 0:
        deleteCols.append(col)
```

```python
colHave = list(df.columns)

smallerCols = [x for x in colHave if x not in deleteCols]
smallerCols
df = df[smallerCols]
df
```

```python
df.to_csv(crimeCSVFILE2, index=False)
```
