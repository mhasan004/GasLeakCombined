```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
csvfile = 'DataFiles/FDNY/Geoid_Count2017_18.csv'
data = pd.read_csv(csvfile)

Season = [['03','04','05'],['06','07','08'],['09','10','11'],['12','01','02']]
Season_word = ['Spring','Summer','Autumn','Winter']
Month_data = pd.DataFrame(columns = ['Geoid','Spring','Summer','Autumn','Winter']) # create a new Dataframe with each month
info = np.array(['36005046201','36047091000'])
Spring_Count = np.array([0,0])
Summer_Count = np.array([0,0])
Autumn_Count = np.array([0,0])
Winter_Count = np.array([0,0])

for row in range(0,len(data)): 
        
    detail = str(data.iat[row,1])
                     
    if detail in info:                          # if the string already exist, we will just increment the count
        itemindex = np.where(info == detail)
    else:                                       # else we will have to add the string to info array and create an extra count
        info = np.append(info,detail)
        Spring_Count = np.append(Spring_Count,0)
        Summer_Count = np.append(Summer_Count,0)
        Autumn_Count = np.append(Autumn_Count,0)
        Winter_Count = np.append(Winter_Count,0)
        itemindex = np.where(info == detail)
        
    if data.iat[row,0][0:2] == '03' or data.iat[row,0][0:2] == '04' or data.iat[row,0][0:2] == '05':
        Spring_Count[itemindex[0][0]] += data.iat[row,2]
    elif data.iat[row,0][0:2] == '06' or data.iat[row,0][0:2] == '07' or data.iat[row,0][0:2] == '08':
        Summer_Count[itemindex[0][0]] += data.iat[row,2]
    elif data.iat[row,0][0:2] == '09' or data.iat[row,0][0:2] == '10' or data.iat[row,0][0:2] == '11':
        Autumn_Count[itemindex[0][0]] += data.iat[row,2]
    elif data.iat[row,0][0:2] == '12' or data.iat[row,0][0:2] == '01' or data.iat[row,0][0:2] == '02':
        Winter_Count[itemindex[0][0]] += data.iat[row,2]
    
for row in range(0,len(info)):
    Month_data = Month_data.append({'Geoid' : info[row], 'Spring' : Spring_Count[row],  'Summer': Summer_Count[row], 'Autumn' : Autumn_Count[row], 'Winter' : Winter_Count[row] } , ignore_index=True)
print(Month_data)
```

```python
Month_data.to_csv('DataFiles/FDNY/Season2017_18.csv', encoding='utf-8',index=False)
```

```python
demo = pd.read_csv("DataFiles/FDNY/Demographic.csv")
join = demo.merge(df,on = "Geoid")
print(join)
```

```python
join['total'] = join['Spring']+join['Summer']+join['Autumn']+join['Winter']
join['popu'] = join['Population'].str.replace(',', '')
print(join)
```

```python
join['total'] = join.total.astype(float)
join['popu'] = join.popu.astype(float)
join['capita'] = join['total']/join['popu']
capita = join.capita.values
join['Povertyrate'] = join['Povertyrate'].str.replace('%', '')
join['Unemploymentrate'] = join['Unemploymentrate'].str.replace('%', '')
join['Povertyrate'] = join.Povertyrate.astype(float)
join['Unemploymentrate'] = join.Unemploymentrate.astype(float)
povertyrate = join.Povertyrate.values
unemploymentrate = join.Unemploymentrate.values
```

```python
print('GRAPHS WITH OUTLIERS')

plt.scatter(capita,povertyrate, color='Green',alpha = 0.1)
plt.xlabel('Number of Gas Leak report per person for census tract')
plt.ylabel('Poverty rate percentage for census tract')
plt.title('Number of Gas Leak Report per person in census tracts of NYC \n vs Poverty rate for census tract in NYC')
plt.show()

plt.scatter(capita,unemploymentrate, color='Green',alpha = 0.1)
plt.xlabel('Number of Gas Leak report per person for census tract')
plt.ylabel('Unemployment rate percentage for census tract')
plt.title('Number of Gas Leak Report per person in census tracts of NYC \n vs Unemployment rate for census tract in NYC')
plt.show()

tot = join.total.values
popu = join.popu.values
plt.scatter(popu,tot, color='Green',alpha = 0.1)
plt.ylabel('Number of Gas Leak report per census tract')
plt.xlabel('Population per census tract')
plt.title('Number of Gas Leak Report per census tracts of NYC \n vs Population per census tract in NYC')
plt.show()
```
