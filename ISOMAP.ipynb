```python
import os
import pandas as pd
import geopandas as gpd
import platform
import matplotlib.pyplot as plt
import contextily as ctx
from tabulate import tabulate
import numpy as np
import matplotlib.colors as mcolors
import PIL.Image as img
from sklearn import manifold
```

```python
res = pd.read_csv("DataFiles/FDNY/ISOMAP_data2017.csv")
print(res)
```

```python
df = res[["Povertyrate","Unemploymentrate","total","popu"]].copy()
iso = manifold.Isomap(n_neighbors=10, n_components=2)
iso.fit(df)
manifold_2Da = iso.transform(df)
manifold_2D = pd.DataFrame(manifold_2Da, columns=['Component1', 'Component2'])

comp1 = manifold_2D.Component1.values
comp2 = manifold_2D.Component2.values
color = res.borough.values
colormap = np.array(['r',    'g',    'b',  'yellow', 'purple'])
#                   bronx, richmond, kings, new york, queen
#360050,360850,360470,360471,360610,360810


plt.scatter(comp1,comp2, c = colormap[color],alpha = 0.6)

```
