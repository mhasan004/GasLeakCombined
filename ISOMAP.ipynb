```python
import os
import pandas as pd
import geopandas as gp
import platform
import matplotlib.pyplot as plt
import contextily as ctx
from tabulate import tabulate
import numpy as np
import matplotlib.colors as mcolors
import PIL.Image as img
from sklearn import manifold
import plotly.express as px
```

```python
res = pd.read_csv("DataFiles/FDNY/ISOMAP_data2017.csv")
crime = pd.read_csv("DataFiles/FDNY/NYPD_Crime.csv")
print(crime)
```

```python
tmax = 0
tmin = 0

for row in range(0,len(crime)):
    if crime.at[row,'TotalCrime'] < tmin:
        tmin = crime.at[row,'TotalCrime']
    elif crime.at[row,'TotalCrime'] > tmax:
        tmax = crime.at[row,'TotalCrime']
        
crime['Crimerate'] = (crime['TotalCrime'] - tmin)/(tmax - tmin)

df2 = crime[["Geoid","Crimerate"]].copy()
df2["Geoid"] = df2.Geoid.astype(int)

res["Geoid"] = res.Geoid.astype(str)
for row in range(0,len(res)):
    res.at[row,"Geoid"] = res.iat[row,0][0:11]

res["Geoid"] = res.Geoid.astype(int)

newdf = res.merge(df2, on = "Geoid")
print(newdf)
```

```python
df = newdf[["Povertyrate","Unemploymentrate","total","popu","Crimerate"]].copy()

iso = manifold.Isomap(n_neighbors=10, n_components=2)
iso.fit(df)
manifold_2Da = iso.transform(df)
manifold_2D = pd.DataFrame(manifold_2Da, columns=['Component1', 'Component2'])

comp1 = manifold_2D.Component1.values
comp2 = manifold_2D.Component2.values
manifold_2D["Geoid"] = newdf.Geoid.values
color = newdf.borough.values
colormap = np.array(['r',    'g',    'b',  'yellow', 'purple'])
boro =  np.array(['Bronx','Staten Island', 'Brooklyn', 'Manhatton', 'Queens'])
#360050,360850,360470,360471,360610,360810

#plt.scatter(comp1,comp2, c = colormap[color],alpha = 0.6)

fig = px.scatter(manifold_2D, x="Component1", y="Component2", color = boro[color],hover_data=['Geoid'])
fig.show()
```
