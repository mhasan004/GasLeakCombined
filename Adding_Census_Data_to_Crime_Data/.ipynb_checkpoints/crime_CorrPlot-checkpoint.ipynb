```python
# What i did for far for the Crime Data:
    # 1) Downlaoded about 200,000 crime reports
    # 2) Managed to only get census data (geoids) of 19,320 reports (was too slow) BUT INCLUDED OTHER BOROUGHS
    # 3) combining all the crime for each geoid turned it into 2092 rows (in the future download for only the 5 booughs)

import pandas as pd
import censusdata
import numpy as np
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.precision', 2)
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt #non interactive scatte rplot


from scipy import stats
import plotly.express as px            #plotting

import plotly.graph_objects as go    # dash sctter (better)
        
```

```python
sesonalFile    = "Season2017_18.csv"
crimeFile      = "NYPD_Complaint_Data_Historic_WITH_CENSUS_DATA_Filtered.csv"
outFile        = "NYPD_Complaint_Data_Historic_WITH_CENSUS_DATA_Filtered_Corr.csv"
seasons        = ["Spring","Summer","Autumn","Winter"]
season_to_show = "ALL"

crimeDF     = pd.read_csv(crimeFile)
seasonalDF  = pd.read_csv(sesonalFile)
crimeDF
```

```python
# Sum up the count of the seasonal report counts and make a new col:
allSum = []
for row in range(0, len(seasonalDF)):
    sum = 0
    for season in seasons:
        sum = sum + seasonalDF.iloc[row][season]
    allSum.append(sum)
seasonalDF["ALL_SEASONS"] = allSum
seasonalDF
```

```python
# 1) Merge (****output csv here)
mainHeaders = ["Geoid", "Spring", "Summer", "Autumn", "Winter", "ALL_SEASONS", "TotalCrime"]
GEOID_COL = "Geoid"
seasonalCrimeDF = pd.merge(seasonalDF, crimeDF, on=GEOID_COL)#, how='left') 
# seasonalCrimeDF = seasonalCrimeDF.filter(mainHeaders)####################################################### can del
seasonalCrimeDF
```

```python

# 2) Get rid of nan rows:
skipIndex = [] 
for row in range(0,len(seasonalCrimeDF)):
    if row in skipIndex:
        continue
    withoutNanN = seasonalCrimeDF.loc[   
        np.isnan(seasonalCrimeDF['TotalCrime'])==False
    ] 
    skipIndex.extend(withoutNanN.index.tolist())  
    withoutNanN = withoutNanN.reset_index(drop=True)
    seasonalCrimeDF = withoutNanN
seasonalCrimeDF = seasonalCrimeDF.sort_values(by=['Geoid'], ascending=False)
seasonalCrimeDF = seasonalCrimeDF.reset_index(drop=True)
seasonalCrimeDF
```

```python
# from scipy import stats # For in-built method to get PCC

# pearson_coef, p_value = stats.pearsonr(seasonalCrimeDF)["ALL_SEASONS"], seasonalCrimeDF["TotalCrime"]) #define the columns to perform calculations on
# print(pearson_coef)
# print(p_value)
```

```python
# 4) PEARSON R CORRELATION:
corr = seasonalCrimeDF.corr(method='pearson')
corr.index.name = 'Crimes'                          # name the index column
corr = corr.reset_index()                                        # numbering the rows and setting DemographicVariable as a new coumn
corr
```

```python
# 5) Getting rid of the repeated season rows (they are repeated in the columns): (***output csv after corr)
for season in seasons:
    corr = corr[corr.Crimes != season] 
corr = corr[corr.Crimes != "ALL_SEASONS"] 
corr = corr[corr.Crimes != "Geoid"] 


# Need to delete the repeated demographic variable columns (they are in the rows already)
delHeader = list(corr.columns)
for i in range(0, len(seasons)):
    delHeader.remove(seasons[i])
delHeader.remove("Crimes")
delHeader.remove("ALL_SEASONS")

count = 0
for i in range(0, len(delHeader)):
    try:
        corr = corr.drop(delHeader[i], axis=1)
    except:
        count = count + 1
        print(str(count)+ " - couldn't drop: "+delHeader[i]+"\n")
corr = corr.reset_index()
corr = corr.drop(['index'], axis=1)              # i have two indexes, so droping old one
# corr.to_csv(outFile, index=False)
corr
```
