```python
# Same graph as shown in plotly_season.ipynb before redone in plotly(which is better).



import os
import pandas as pd
import geopandas as gp
import platform
import matplotlib.pyplot as plt
import contextily as ctx
from tabulate import tabulate
import numpy as np
import matplotlib.colors as mcolors
import PIL.Image as img
from sklearn import manifold
import plotly.express as px
```

```python
Month_data = pd.read_csv('DataFiles/FDNY/Season2017_18.csv')

Month_data['borough'] = 0

#360050,360850,360470,360471,360610,360810
#  bronx,richmond,kings,new york,queen
for row in range(0,len(Month_data)):
    if int(Month_data.at[row,'Geoid']/100000) == 360050:
        Month_data.at[row,'borough'] = 0
    elif int(Month_data.at[row,'Geoid']/100000) == 360850:
        Month_data.at[row,'borough'] = 1
    elif int(Month_data.at[row,'Geoid']/100000) == 360470 or int(Month_data.at[row,'Geoid']/100000) == 360471:
        Month_data.at[row,'borough'] = 2
    elif int(Month_data.at[row,'Geoid']/100000) == 360610:
        Month_data.at[row,'borough'] = 3
    elif int(Month_data.at[row,'Geoid']/100000) == 360810:
        Month_data.at[row,'borough'] = 4

color = Month_data.borough.values
boro =  np.array(['Bronx','Staten Island', 'Brooklyn', 'Manhatton', 'Queens'])

px.scatter(Month_data, x="Spring", y="Summer", color = boro[color],hover_data=['Geoid'],title = 'Number of Gas Leak Report per census tract in NYC (Spring vs Summer)')
```

```python
px.scatter(Month_data, x="Summer", y="Autumn", color = boro[color],hover_data=['Geoid'],title = 'Number of Gas Leak Report per census tract in NYC (Summer vs Autumn)')
```

```python
px.scatter(Month_data, x="Autumn", y="Winter", color = boro[color],hover_data=['Geoid'],title = 'Number of Gas Leak Report per census tract in NYC (Autumn vs Winter)')
```

```python
px.scatter(Month_data, x="Winter", y="Spring", color = boro[color],hover_data=['Geoid'],title = 'Number of Gas Leak Report per census tract in NYC (Winter vs Spring)')
```

```python
demo = pd.read_csv("DataFiles/FDNY/Demographic.csv")
join = demo.merge(Month_data,on = "Geoid")
join['total'] = join['Spring']+join['Summer']+join['Autumn']+join['Winter']
join['popu'] = join['Population'].str.replace(',', '')
join['total'] = join.total.astype(float)
join['popu'] = join.popu.astype(float)
join['capita'] = join['total']/join['popu']
join['Povertyrate'] = join['Povertyrate'].str.replace('%', '')
join['Unemploymentrate'] = join['Unemploymentrate'].str.replace('%', '')
join['Povertyrate'] = join.Povertyrate.astype(float)
join['Unemploymentrate'] = join.Unemploymentrate.astype(float)

for row in range(0,len(join)):
    if int(join.at[row,'Geoid']/100000) == 360050:
        join.at[row,'borough'] = 0
    elif int(join.at[row,'Geoid']/100000) == 360850:
        join.at[row,'borough'] = 1
    elif int(join.at[row,'Geoid']/100000) == 360470 or int(join.at[row,'Geoid']/100000) == 360471:
        join.at[row,'borough'] = 2
    elif int(join.at[row,'Geoid']/100000) == 360610:
        join.at[row,'borough'] = 3
    elif int(join.at[row,'Geoid']/100000) == 360810:
        join.at[row,'borough'] = 4

color2 = join.borough.values

px.scatter(join, x="Povertyrate", y="capita", color = boro[color2],hover_data=['Geoid'],title = 'Poverty rate per CT vs number of gas leak report per person in CT')
```

```python
px.scatter(join, x="Unemploymentrate", y="capita", color = boro[color2],hover_data=['Geoid'],title = 'Unemployment rate per CT vs number of gas leak report per person in CT')
```

```python
px.scatter(join, x="popu", y="total", color = boro[color2],hover_data=['Geoid'],title = 'Population per CT vs Total Number of Gas Leak report per CT')
```
