```python
# Using censusdata API:  https://pypi.org/project/CensusData/. 
# GOAL - DASHBOARD 3: Adding new census tract data to 2017-2018 season csv and outputing to file. dashboard3.py will import file and run it in dash
import pandas as pd
import censusdata
import numpy as np
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.precision', 2)
import statsmodels.formula.api as sm
```

```python
# PARAMETER OF SEARCH
searchCriteria = 'percentage'
startEndRow = [0,300]
seasonalCSVFile = "DataFiles/FDNY/Seasonal2017_2018_Demographics.csv"
```

```python
 # 1) Find a table by looking at the variables of all tables
table = censusdata.search('acs5', 2018, 'concept', searchCriteria)[startEndRow[0]:startEndRow[1]]
tableColDict = dict()
tableColCodeArray = []
for i in range(0,len(table)):  
    tableColCode = str(table[i]).split(", ")[0].strip("(").strip(")").strip("'")
    tableColSection = str(table[i]).split(", ")[1].strip("(").strip(")").strip("'")
    tableColVar = str(table[i]).split(", ")[2].strip("(").strip(")").strip("'")
    
    tableColDict.update( {tableColCode: [tableColSection,tableColVar]} )
    tableColCodeArray.append(tableColCode)
count = 0
for key in tableColDict:
#     print(str(count)+  "..."+ 
    print(key + "\n              " + tableColDict[key][0] +  "\n              " + tableColDict[key][1])
    count = count + 1
    if count == 101:
        break
keys = list(tableColDict.keys())
if (len(table)<=0):
    print("RETRIEVED NOTHING INPUT NAOTHER SEARCH CRITERIA")
    exit()
```

```python
# Merge into GEOID col and make a CountyName Col
df["CountyName"] = np.str 

for row in range(0, len(df)):
    splitTractName = str(df.iloc[row]["Geoid"]).split(',')

    countyName = splitTractName[1]
    
    geoidSplit = splitTractName[3].split("> ")
    geoid = int(geoidSplit[0].split(':')[1] + geoidSplit[1].split(':')[1] + geoidSplit[2].split(':')[1] )
    
    df.at[row, 'Geoid'] = geoid
    df.at[row, 'CountyName'] = countyName
df
```
